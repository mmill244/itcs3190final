# I will be transforming the data to the get the average home pricing for each state

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading in the csv
zillow_df = pd.read_csv('zillow.csv')
zillow_df
```

```python
#transformation 1 - removing unnecessary columns
zillow_df = zillow_df.drop(['SizeRank', 'RegionID', 'RegionName', 'RegionType'], 1)
zillow_df
```

```python
#transformation 2 - removing rows with NaN values
zillow_df = zillow_df.dropna(axis=0)
zillow_df
```

```python
#transformation 3 - sorting column "StateName" into ascending order
zillow_df = zillow_df.sort_values("StateName", ascending=True)
zillow_df
```

```python
#transformation 4 - getting the average of all home prices for each row
zillow_df['mean'] = zillow_df.mean(axis=1)
zillow_df
```

```python
#transformation 5 - dropping all columns with home prices
zillow_df.drop(zillow_df.iloc[:, 1:263], axis=1, inplace=True)
zillow_df
```

```python
#transformation 6 - getting the average home price for each state
zillow_df.groupby(['StateName'], as_index=False).mean()
```
